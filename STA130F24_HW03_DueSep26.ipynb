````python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
````

````python
# Calculate statistics for each species
summary_stats = penguins.groupby('species')['flipper_length_mm'].agg(['mean', 'median', 'min', 'max', 'std', 'quantile']).reset_index()

# Calculate interquartile range (IQR)
summary_stats['IQR'] = penguins.groupby('species')['flipper_length_mm'].apply(lambda x: x.quantile(0.75) - x.quantile(0.25)).values

# Calculate range ± 2σ from the mean
summary_stats['lower_2sigma'] = summary_stats['mean'] - 2 * summary_stats['std']
summary_stats['upper_2sigma'] = summary_stats['mean'] + 2 * summary_stats['std']
````

````python
# Initialize figure
fig = go.Figure()

# Add histogram and statistics lines for each species
for species in summary_stats['species'].unique():
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
    # Add histogram
    fig.add_trace(go.Histogram(x=species_data, name=species, opacity=0.7))
    
    # Get statistics for the current species
    stats = summary_stats[summary_stats['species'] == species]
    
    # Add lines for mean, median, and ±2σ range
    fig.add_vline(x=stats['mean'].values[0], line=dict(color='red', dash='dash'), annotation_text='Mean', annotation_position="top left")
    fig.add_vline(x=stats['median'].values[0], line=dict(color='blue', dash='dash'), annotation_text='Median', annotation_position="top right")
    fig.add_vrect(x0=stats['lower_2sigma'].values[0], x1=stats['upper_2sigma'].values[0], fillcolor='green', opacity=0.2, annotation_text='±2σ Range', annotation_position="top left")

fig.update_layout(title='Flipper Length Distribution by Species',
                  xaxis_title='Flipper Length (mm)',
                  yaxis_title='Count',
                  barmode='overlay',
                  template='plotly_white')

fig.show(renderer="png")
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
````

````python
# Calculate summary statistics for each species
summary_stats = penguins.groupby('species')['flipper_length_mm'].agg(['mean', 'median', 'min', 'max', 'std']).reset_index()

# Calculate interquartile range (IQR)
summary_stats['IQR'] = penguins.groupby('species')['flipper_length_mm'].apply(lambda x: x.quantile(0.75) - x.quantile(0.25)).values

# Calculate range ± 2σ from the mean
summary_stats['lower_2sigma'] = summary_stats['mean'] - 2 * summary_stats['std']
summary_stats['upper_2sigma'] = summary_stats['mean'] + 2 * summary_stats['std']
````

````python
# Set up the plot grid
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharex=True, sharey=True)
species_list = summary_stats['species'].unique()

# Plot KDE for each species
for i, species in enumerate(species_list):
    ax = axes[i]
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
    # Plot KDE
    sns.kdeplot(data=species_data, ax=ax, fill=True, color='skyblue', label=f'{species} KDE')
    
    # Get statistics for the current species
    stats = summary_stats[summary_stats['species'] == species].iloc[0]
    
    # Add vertical lines for mean and median
    ax.axvline(stats['mean'], color='red', linestyle='--', label=f'Mean: {stats["mean"]:.2f}')
    ax.axvline(stats['median'], color='blue', linestyle='--', label=f'Median: {stats["median"]:.2f}')
    
    # Add shaded area for ±2σ range
    ax.axvspan(stats['lower_2sigma'], stats['upper_2sigma'], color='green', alpha=0.2, label='±2σ Range')
    
    # Set titles and labels
    ax.set_title(f'Flipper Length Distribution for {species}')
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Density')
    ax.legend()
    
# Adjust layout for clarity
plt.tight_layout()
plt.savefig("my_plot.png")
plt.show()
````

Here's the pros and cos:

1. Box Plots

Pros:
Summarizes key statistics: Provides a compact representation of the minimum, 25th percentile (Q1), median, 75th percentile (Q3), and maximum values.

Cons:

Limited detail: Only provides summary statistics, so it may hide subtle patterns such as bimodal distributions or heavy tails.
No information on frequency: Doesn’t give insight into how frequently different values occur​(
Modeling Life
).

2. Histograms

Pros:
Visualizes frequency: Directly shows the frequency of data within defined intervals (bins), giving a good sense of the distribution.
Intuitive to interpret: Easy to understand and widely used in introductory data analysis.
Customizable: By changing the bin width, you can adjust how detailed the visualization is, allowing you to zoom in on fine features or show an overview.

Cons:
Sensitive to bin size: The shape of the histogram can vary significantly depending on the bin width. Too many bins can show random noise, while too few bins can oversimplify the data.
Not smooth: Unlike KDE, histograms can appear “blocky” due to the binning process.
Difficult to compare: Comparing histograms across multiple datasets can be tricky unless the bin widths and cutoffs are consistent​(
Modeling Life
)​(
AstroML

3. Kernel Density Estimators (KDE)

Pros:
Smooths data: Provides a continuous representation of the data distribution, offering a clearer view of the shape compared to histograms.
Reveals subtle features: KDEs are good at highlighting fine details in the data, such as multiple peaks (multimodality) or asymmetry.
Flexible: The smoothness can be adjusted by changing the bandwidth, which controls how much each data point influences its neighbors.

Cons:
Sensitive to bandwidth: The shape of the KDE can be significantly influenced by the bandwidth parameter, potentially obscuring true features if the bandwidth is not chosen carefully.

````python
from scipy import stats
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

n = 1500
data1 = stats.uniform.rvs(0, 10, size=n)
data2 = stats.norm.rvs(5, 1.5, size=n)
data3 = np.r_[stats.norm.rvs(2, 0.25, size=int(n/2)), stats.norm.rvs(8, 0.5, size=int(n/2))]
data4 = stats.norm.rvs(6, 0.5, size=n)

fig = make_subplots(rows=1, cols=4)

fig.add_trace(go.Histogram(x=data1, name='A', nbinsx=30, marker=dict(line=dict(color='black', width=1))), row=1, col=1)
fig.add_trace(go.Histogram(x=data2, name='B', nbinsx=15, marker=dict(line=dict(color='black', width=1))), row=1, col=2)
fig.add_trace(go.Histogram(x=data3, name='C', nbinsx=45, marker=dict(line=dict(color='black', width=1))), row=1, col=3)
fig.add_trace(go.Histogram(x=data4, name='D', nbinsx=15, marker=dict(line=dict(color='black', width=1))), row=1, col=4)

fig.update_layout(height=300, width=750, title_text="Row of Histograms")
fig.update_xaxes(title_text="A", row=1, col=1)
fig.update_xaxes(title_text="B", row=1, col=2)
fig.update_xaxes(title_text="C", row=1, col=3)
fig.update_xaxes(title_text="D", row=1, col=4)
fig.update_xaxes(range=[-0.5, 10.5])

for trace in fig.data:
    trace.xbins = dict(start=0, end=10)
    
# This code was produced by just making requests to Microsoft Copilot
# https://github.com/pointOfive/stat130chat130/blob/main/CHATLOG/wk3/COP/SLS/0001_concise_makeAplotV1.md

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

Mean is the average of a set of numbers.

Median is the middle value of a dataset when ordered from minimum to maximum.

Skewness is a measure of the asymmetry of a distribution.

Right Skew is the tail on the right side is longer or fatter than the left side.

Left Skew is the tail on the left side is longer or fatter than the right side.

When the mean is less than the median, it is often an indication that the data distribution is left-skewed. In a left-skewed distribution, the data has long tails on the left side, meaning there are extremely small values that pull the mean down.

When the mean is greater than the median, it is often an indication that the data distribution is right-skewed. In a right-skewed distribution, the data has long tails to the right, which means that there are some extremely large values that pull the mean up.

````python
import pandas as pd
import plotly.express as px

# Load the baby names dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Create a unique identifier for names by appending gender
bn['name'] = bn['name'] + " " + bn['sex']  # Make identical boy and girl names distinct

# Rank names within each year based on their prevalence
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)

# Sort values for consistency
bn = bn.sort_values(['name', 'year'])

# Calculate the percentage change in name prevalence from the previous year
bn['percent change'] = bn['percent'].diff()

# Identify new names to reset the percentage change
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']

# Further sort values by year
bn = bn.sort_values('year')

# Filter out uncommon names with very low prevalence
bn = bn[bn.percent > 0.001]  # Keep only "common" names

# Create a scatter plot to visualize name trends over the years
fig = px.scatter(
    bn,
    x="percent change",
    y="rank",
    animation_frame="year",
    animation_group="name",
    size="percent",
    color="sex",
    hover_name="name",
    size_max=50,
    range_x=[-0.005, 0.005]  # X-axis range for clarity
)

# Reverse the Y-axis to display the highest rank at the top
fig.update_yaxes(autorange='reversed')

# Show the figure in PNG format for submissions
fig.show(renderer="png")
````
